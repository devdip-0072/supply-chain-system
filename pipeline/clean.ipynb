```python
import pandas as pd

df = pd.read_csv(
    r'D:\Supply_chain_project\data\raw\table.csv',
    dtype={
        'promotion_type': str,
        'channel': str
    }
)
```

```python
# See how many missing values are in each column
missing = df.isnull().sum()
missing_percentage = (missing / len(df)) * 100
missing_df = pd.DataFrame({'Missing Values': missing, 'Percentage (%)': missing_percentage})
missing_df[missing_df['Missing Values'] > 0].sort_values(by='Percentage (%)', ascending=False)
```

```python
# Drop columns with more than 90% missing values
df.drop(columns=['social_media_mentions', 'weather_condition', 'temperature', 'competitor_analysis_score'], inplace=True)
```

```python
for i, col in enumerate(df.columns):
    print(f"{i}: {col}")
```

```python
# Fill missing values in categorical columns with the mode (no inplace=True)
df['channel'] = df['channel'].fillna(df['channel'].mode()[0])
df['promotion_type'] = df['promotion_type'].fillna(df['promotion_type'].mode()[0])

# Fill missing values in numerical columns with the mean
df['discount_percentage'] = df['discount_percentage'].fillna(df['discount_percentage'].mean())
```

```python
# Check missing values after filling
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100
print(f"Missing Values after cleaning:\n{missing_values}")
print(f"Missing Values Percentage:\n{missing_percentage}")
```
